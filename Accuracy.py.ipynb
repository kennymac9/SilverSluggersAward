```python
# cols = ['runs', 'HRs', 'RBIs', 'AVG', 'Ops']
actualCols = ['runs', 'hr', 'rbi', 'avg', 'ops']
predictedCols = ['runsPredicted', 'HRsPredicted', 'RBIsPredicted', 'AVGPredicted', 'opsPredicted']
# predictedCols = ['linearSVRRuns', 'linearSvrHR', 'linearSvrRBI', 'linearSvrAVG', 'linearSvrOPS']
# predictedCols = ['svrRuns', 'SvrHR', 'svrRBI', 'svrAVG', 'svrOPS']
# predictedCols = ['HuberRegressorRuns', 'HuberRegressorHR', 'HuberRegressorRBI', 'HuberRegressorAVG', 'HuberRegressorOPS']
# predictedCols = ['RidgeRuns', 'RidgeHR', 'RidgeRBI', 'RidgeAVG', 'RidgeOPS', ]
mae = {}
rmse = {}
```

```python
files = ['allPredictions2017.csv', 'allPredictions2018.csv', 'allPredictions2019.csv']
maeObj = { 'runs': [], 'hr': [], 'rbi': [], 'avg': [], 'ops': []}
rmseObj = { 'runs': [], 'hr': [], 'rbi': [], 'avg': [], 'ops': []}
for file in files:
    df = pd.read_csv(file, index_col=0)
    ndf = df.dropna()

    # cols = ['runs', 'HRs', 'RBIs', 'AVG', 'Ops']
    actualCols = ['runs', 'hr', 'rbi', 'avg', 'ops']
#     predictedCols = ['runsPredicted', 'HRsPredicted', 'RBIsPredicted', 'AVGPredicted', 'opsPredicted']
#     predictedCols = ['linearSVRRuns', 'linearSvrHR', 'linearSvrRBI', 'linearSvrAVG', 'linearSvrOPS']
#     predictedCols = ['svrRuns', 'SvrHR', 'svrRBI', 'svrAVG', 'svrOPS']
#     predictedCols = ['HuberRegressorRuns', 'HuberRegressorHR', 'HuberRegressorRBI', 'HuberRegressorAVG', 'HuberRegressorOPS']
    predictedCols = ['RidgeRuns', 'RidgeHR', 'RidgeRBI', 'RidgeAVG', 'RidgeOPS', ]
    for i, col in enumerate(actualCols):
        print(col)
        predCol = predictedCols[i]
        maeObj[col].append(mean_absolute_error(ndf[col], ndf[predCol]))
        rmseObj[col].append(math.sqrt(mean_squared_error(ndf[col], ndf[predCol])))

    
```

```python
for cat in maeObj:
    print(sum(maeObj[cat]) / 3)
```

```python
for cat in rmseObj:
    print(sum(rmseObj[cat]) / 3)
```
